dataset = {'Group': ['G-2', 'G-3', 'G-3', 'G-2', 'G-2',

                    '2020-12-25', '2018-06-01', '2018-07-15',
                    '2019-09-14']}
  
dataset = pd.DataFrame(dataset, columns=['Group', 'Date'])
  
# using groupby() function on Group column
df = dataset.groupby(['Group'])
  
# using agg() function on Date column
df2 = df.agg(Minimum_Date=('Date', np.min), Maximum_Date=('Date', np.max))
  
# Displaying result
display(df2)

```python
#Top 10 causes of death for Afghanistan in 2019 (latest available year)

afg_2019 = df[mask].groupby("Year").sum().loc[2019].sort_values(ascending=False)
sns.barplot(x=afg_2019.values[:15],y=afg_2019.index[:15],orient="h")
```

```python
df_c = df.groupby(['Country/Territory'])
```

```python
df_c.sum()['Malaria'].sort_values(ascending=False)
```

```python
#Top 20 countries which suffered from malaria the most.

cntry_malaria = df_c.sum()['Malaria'].sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=cntry_malaria[1:21].values,y=cntry_malaria[1:21].index,orient="h")
plt.xlabel("Total Deaths 1990-2019")
```

```python
df["total_death"]=df.iloc[:,3:].sum(axis=1)
print(df["total_death"])
death_by_year = df.groupby("Year").sum()["total_death"]
fig = plt.figure(figsize=(9,7))
death_by_year.plot(marker="o")
plt.ylabel("# of Deaths")
```

```python
#Deadliest Year

df["total_death"] = df.iloc[:,3:].sum(axis=1)
print(df["total_death"])
death_by_year = df.groupby("Year").sum()["total_death"]
fig = plt.figure(figsize=(9,7))
death_by_year.plot(marker="o")
plt.ylabel("# of Deaths")
```
