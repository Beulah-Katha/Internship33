# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

Note: All the steps required during scraping should be done through code only and not manually.

```python

# Get the webpage
url =  'https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART'
#url = 'https://www.https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART'
driver.get(url)
```

```python

# scraping ratings, review summary and full review

rates = [] 
review_summary = []
full_review = []

page_start = 0
page_end = 10

for page in range(page_start,page_end):
    
    rates_tag = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rates_tag:
        rates.append(i.text)
    
    summary_tag = driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for j in summary_tag:
        review_summary.append(j.text)

    review_tag = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for k in review_tag:
        full_review.append(k.text)    
    
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]//span')
    next_button.click()
    time.sleep(3)
    
```

```python
# creating dataframe

iphone_df = pd.DataFrame( {'Ratings':rates,
                      'Review Summary':review_summary,
                      'Full Review':full_review                       
                          } )
iphone_df
```

# Q10: Write s python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpage https://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make the DataFrame.

```python
# opening amazon website

url = 'https://www.jagranjosh.com/'
driver.get(url)
```

```python
# click GK option

gk_button = driver.find_element(By.XPATH,'/html/body/div/div[1]/div/div[1]/div/div[6]/div/div[1]/header/div[3]/ul/li[9]/a')
gk_button.click()
```

```python
# click Prime Minister option

pm_button = driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a')
pm_button.click()
```

```python
row_count = len(driver.find_elements(By.XPATH,'//*[@id="itemdiv"]/div[5]/span/div[2]/table/tbody/tr'))
#col_count = len(driver.find_elements(By.XPATH,'//*[@id="itemdiv"]/div[5]/span/div[2]/table/tbody/tr[1]/th'))
#print(row_count)
#print(col_count)

names = []
born_dead = []
office_term = []
remarks = []

for r in range(2, row_count+1):
    names_columns = driver.find_elements(By.XPATH,'//*[@id="itemdiv"]/div[5]/span/div[2]/table/tbody/tr["+str(r)"]/td[2]')
    bd_columns = driver.find_elements(By.XPATH,'//*[@id="itemdiv"]/div[5]/span/div[2]/table/tbody/tr["+str(r)"]/td[3]')
    ot_columns = driver.find_elements(By.XPATH,'//*[@id="itemdiv"]/div[5]/span/div[2]/table/tbody/tr["+str(r)"]/td[4]')
    rem_columns = driver.find_elements(By.XPATH,'//*[@id="itemdiv"]/div[5]/span/div[2]/table/tbody/tr["+str(r)"]/td[5]')
    
for c1 in names_columns:
    names.append(c1.text)
    
for c2 in bd_columns:
    born_dead.append(c2.text)
    
for c3 in ot_columns:
    office_term.append(c3.text)
    
for c4 in rem_columns:
    remarks.append(c4.text)
```

```python
# creating dataframe

pm_df = pd.DataFrame( {'Name':names,
                        'Born-Dead':born_dead,
                        'Term of Office':office_term,
                       'Remarks':remarks
                             } )
pm_df
```

```python
exp50cars = driver.find_element(By.XPATH,'/html/body/div[2]/div[9]/div[1]/div[1]/div/div/div[1]/div/div[1]/h3/a')
exp50cars.click()
#/html/body/div[2]/div[9]/div[1]/div[1]/div/div/div[1]/div/div[1]/h3/a
```

```python
# Extracting Model name

model_names = []
prices = []


model_tags = driver.find_elements(By.XPATH,'//h3[@class="subheader"]')

price_tags = driver.find_elements(By.XPATH,'//div[@class="postBody description e-content"]//p//strong')

for i in model_tags:
    model_names.append(i.text)

for j in price_tags:
    prices.append(j.text)
```
